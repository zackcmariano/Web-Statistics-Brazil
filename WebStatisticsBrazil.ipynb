### Em um ano que o planeta passou por muitas mudanças, a tecnologia foi potencializada por bilhões de pessoas confinadas em suas casas e contando com soluções a um click de distância. E, em meio ao afastamento interpessoal, a tecnologia se aproximou e criou raízes nos lugares mais improváveis. 
### Na história da humanidade, nunca se viu um cruzamento de dados tão grande e tão eficientes quanto tem sido nos últimos meses.
###Dados estatísticos da gigante Irlandesa StatCounter, com 22 anos de mercado, que usa sua tecnologia para calcular o compartilhamento de uso da web, mostram grandes mudanças nos hábitos digitais dos brasileiros.
###Os dados da StatCounter são derivados diretamente dos acessos de 3 milhões de sites, resultando em um total de mais de 15 bilhões de acessos por mês. 
###Não são utilizadas ponderações artificiais para corrigir o enviesamento da amostragem, sendo assim, os números nas estatísticas não podem ser considerados amostras representativas.

```
#USED LIBRARIES

import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import pyplot as pls
import numpy as np
%matplotlib inline

plt.style.use("ggplot")
```

```
#IMPORT FILES REQUIRED
operational_system = pd.read_csv('https://raw.githubusercontent.com/zackcmariano/Web-Statistics-Brazil/main/os_combined-BR-monthly-202008-202101.csv')
browser = pd.read_csv('https://raw.githubusercontent.com/zackcmariano/Web-Statistics-Brazil/main/browser-BR-monthly-202008-202101.csv')
social_media = pd.read_csv('https://raw.githubusercontent.com/zackcmariano/Web-Statistics-Brazil/main/social_media-BR-monthly-202008-202101.csv')
devices = pd.read_csv('https://raw.githubusercontent.com/zackcmariano/Web-Statistics-Brazil/main/comparison-BR-monthly-202008-202101.csv')
```

#SISTEMA OPERACIONAL:
###Embora em maior parte do mundo o Sistema Android é o queridinho da Web aqui no Brasil. O gigante da Microsoft, Windows, se mantém no topo no semestre, mas tem seu primeiro lugar comprometido pelo Android que tomou a posição de primeiro lugar em Dezembro/2020 e Janeiro/2021. E em terceiro lugar temos o iOS

```
data = operational_system.Date
year = ['Ago/20','Set/20','Out/20','Nov/20','Dez/20','Jan/21']
x = np.arange(len(data))

fig,ax = plt.subplots(figsize=(16,6))

windows = ax.bar(x - 0.30, operational_system.Windows, width = 0.30, color = '#23807e')
android = ax.bar(x, operational_system.Android, width = 0.30, color = '#6ae1e2')
ios = ax.bar(x + 0.30, operational_system.iOS, width = 0.30, color = '#2e95b4')

plt.xticks(fontsize=14)
ax.set_title('SISTEMA OPERACIONAL', fontsize=20, color='#1f6376')
ax.set_ylabel('Porcentagem %', fontsize=15, color='#25424a')

ax.set_ylim([0,100])
ax.set_xticks(x)
ax.set_xticklabels(year)

plt.legend(labels=['Windows', 'Android', 'iOS'], fontsize=14);

def autolabel(system):
  for i in system:
    percentage = i.get_height()
    ax.annotate('{}'.format(percentage),
                xy = (i.get_x()+i.get_width()/2,percentage),
                xytext = (0,5),
                textcoords = 'offset points',
                ha = 'center', fontsize=12)    

autolabel(windows)
autolabel(android)
autolabel(ios)

#plt.savefig('BrowsersBr.png',dpi = 600)   / TO SAVE PNG

plt.show()
```

#APARELHOS:
###Existe uma concorrência muito grande entre Desktop e Mobile no Brasil. Embora o Desktop lidere os últimos seis meses, dados mostram que o Mobile está vencendo a corrida nos últimos 60 dias. 
###Quem levará a melhor esse ano? uma vez que o Tablet vem muito atrás, em terceiro lugar, e não representa perigo algum a essa grande disputa.

```
data = devices.Date
year = ['Ago/20','Set/20','Out/20','Nov/20','Dez/20','Jan/21']
x = np.arange(len(data))

fig,ax = plt.subplots(figsize=(16,6))

desktop = ax.bar(x - 0.30, devices.Desktop, width = 0.30, color = '#9b013d')
mobile = ax.bar(x, devices.Mobile, width = 0.30, color = '#d73743')
tablet = ax.bar(x + 0.30, devices.Tablet, width = 0.30, color = '#d10159')

plt.xticks(fontsize=14)
ax.set_title('APARELHOS', fontsize=20, color='#ab0535')
ax.set_ylabel('Porcentagem %', fontsize=15, color='#81001f')

ax.set_ylim([0,100])
ax.set_xticks(x)
ax.set_xticklabels(year)

plt.legend(labels=['Desktop', 'Mobile', 'Tablet'], fontsize=14);

def autolabel(device):
  for i in device:
    percentage = i.get_height()
    ax.annotate('{}'.format(percentage),
                xy = (i.get_x()+i.get_width()/2,percentage),
                xytext = (0,5),
                textcoords = 'offset points',
                ha = 'center', fontsize=12)    

autolabel(desktop)
autolabel(mobile)
autolabel(tablet)

#plt.savefig('BrowsersBr.png',dpi = 600)   / TO SAVE PNG

plt.show()
```

##A Estatística nos mostra que no último semestre algumas das grandes marcas se mantiveram no topo, outras já não fazem mais parte da preferência dos Brasileiros.
##É indiscutível que o mundo realmente mudou nesses últimos meses e que grandes empresas, que antes dominavam o mercado, aos poucos vão deixando o topo e repetindo o mesmo ciclo que já aconteceu com outras gigantes que hoje simplesmente não existem mais.
